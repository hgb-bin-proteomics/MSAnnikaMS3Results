```python
import pandas as pd
from venn import venn

# CID
cid_maxlinker = pd.read_csv("CID/MaXLinker_XLpeplib_Beveridge_Lumos_DSSO_MS3_out_gbSP_output.csv")
cid_ms2annika = pd.read_csv("CID/MS2Annika_XLpeplib_Beveridge_Lumos_DSSO_MS3-(93)_output.csv")
cid_ms3annika = pd.read_csv("CID/MS3Annika_XLpeplib_Beveridge_Lumos_DSSO_MS3-(97)_output.csv")
cid_xlinkx = pd.read_csv("CID/XlinkX_XLpeplib_Beveridge_Lumos_DSSO_MS3-(2)_output.csv")

# EThcD
etd_maxlinker = pd.read_csv("EThcD/MaXLinker_XLpeplib_Beveridge_Lumos_DSSO_MS3-EThcD_out_gbSP_output.csv")
etd_ms2annika = pd.read_csv("EThcD/MS2Annika_XLpeplib_Beveridge_Lumos_DSSO_MS3-EThcD-(201)_output.csv")
etd_ms3annika = pd.read_csv("EThcD/MS3Annika_XLpeplib_Beveridge_Lumos_DSSO_MS3-EThcD-(205)_output.csv")
etd_xlinkx = pd.read_csv("EThcD/XlinkX_XLpeplib_Beveridge_Lumos_DSSO_MS3-EThcD-(2)_output.csv")
```

```python
def get_crosslink(dataset_row):
    seq_a = dataset_row["Sequence A"]
    seq_b = dataset_row["Sequence B"]
    
    return "-".join(sorted([seq_a, seq_b]))
```

```python
cid_maxlinker_xls = set(cid_maxlinker.loc[cid_maxlinker["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
cid_ms2annika_xls = set(cid_ms2annika.loc[cid_ms2annika["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
cid_ms3annika_xls = set(cid_ms3annika.loc[cid_ms3annika["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
cid_xlinkx_xls = set(cid_xlinkx.loc[cid_xlinkx["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))

etd_maxlinker_xls = set(etd_maxlinker.loc[etd_maxlinker["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
etd_ms2annika_xls = set(etd_ms2annika.loc[etd_ms2annika["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
etd_ms3annika_xls = set(etd_ms3annika.loc[etd_ms3annika["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
etd_xlinkx_xls = set(etd_xlinkx.loc[etd_xlinkx["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
```

```python
# colors
# MaXLinker, MS2 Annika, MS3 Annika, XlinkX
col = ["#4361EE", "#7209B7", "#4CC9F0", "#F72585"]
```

```python
venn(cid_ms2annika_xls, cid_ms3annika_xls, labels = ["MS2 Annika", "MS3 Annika"], colors = [col[1], col[2]], contour = True, title = "CID");
```

```python
venn(cid_maxlinker_xls, cid_ms3annika_xls, cid_xlinkx_xls, labels = ["MaXLinker", "MS3 Annika", "XlinkX"], contour = True, title = "CID");
```

```python
venn(etd_ms2annika_xls, etd_ms3annika_xls, labels = ["MS2 Annika", "MS3 Annika"], colors = [col[1], col[2]], contour = True, title = "EThcD");
```

```python
venn(etd_maxlinker_xls, etd_ms3annika_xls, etd_xlinkx_xls, labels = ["MaXLinker", "MS3 Annika", "XlinkX"], contour = True, title = "EThcD");
```
