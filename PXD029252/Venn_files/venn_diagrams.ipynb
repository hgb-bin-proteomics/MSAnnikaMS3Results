```python
def get_crosslink(dataset_row):
    seq_a = dataset_row["Sequence A"]
    seq_b = dataset_row["Sequence B"]
    
    return "-".join(sorted([seq_a, seq_b]))
```

```python
rep1_maxlinker_xls = set(rep1_maxlinker.loc[rep1_maxlinker["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
rep1_ms2annika_xls = set(rep1_ms2annika.loc[rep1_ms2annika["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
rep1_ms3annika_xls = set(rep1_ms3annika.loc[rep1_ms3annika["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
rep1_xlinkx_xls = set(rep1_xlinkx.loc[rep1_xlinkx["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))

rep2_maxlinker_xls = set(rep2_maxlinker.loc[rep2_maxlinker["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
rep2_ms2annika_xls = set(rep2_ms2annika.loc[rep2_ms2annika["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
rep2_ms3annika_xls = set(rep2_ms3annika.loc[rep2_ms3annika["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
rep2_xlinkx_xls = set(rep2_xlinkx.loc[rep2_xlinkx["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))

rep3_maxlinker_xls = set(rep3_maxlinker.loc[rep3_maxlinker["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
rep3_ms2annika_xls = set(rep3_ms2annika.loc[rep3_ms2annika["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
rep3_ms3annika_xls = set(rep3_ms3annika.loc[rep3_ms3annika["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
rep3_xlinkx_xls = set(rep3_xlinkx.loc[rep3_xlinkx["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
```
