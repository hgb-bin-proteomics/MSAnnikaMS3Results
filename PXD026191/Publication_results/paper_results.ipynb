```python
import json
import pandas as pd
```

```python
data_paper = pd.read_excel("nsp7_nsp8_nsp7nsp8-1to1_nsp7nsp8-1to2_red_alk_trypsin_XLMS_CSM.xlsx")
```

```python
nsp8_files = ["202001216_nsp8_trypsin_XL_REP1.raw", "202001216_nsp8_trypsin_XL_REP2.raw", "202001216_nsp8_trypsin_XL_REP3.raw"]
```

```python
nsp8_links = data_paper[data_paper["Spectrum File"].isin(nsp8_files)]
```

```python
nsp8_links
```

```python
def clean_seq(seq):
    valid_AA = ["G", "P", "A", "V", "L", "I", "M", "C", "F", "Y", "W", "H", "K", "R", "Q", "N", "E", "D", "S", "T"]
    clean_seq = ""
    for AA in seq:
        if AA in valid_AA:
            clean_seq += AA
    return clean_seq
```

```python
crosslinks = set()

for i, row in nsp8_links.iterrows():
    crosslink = "-".join(sorted([clean_seq(row["Sequence A"]) + str(row["Crosslinker Position A"]), clean_seq(row["Sequence B"]) + str(row["Crosslinker Position B"])]))
    crosslinks.add(crosslink)

with open("paper_crosslinks_nsp8.json", "w", encoding = "utf-8") as f:
    json.dump({"crosslinks": list(crosslinks)}, f)
    f.close()
```
